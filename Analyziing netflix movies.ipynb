**Analyzing the Duration of 1990s Action Movies on Netflix**

```python
import pandas as pd
import matplotlib.pyplot as plt

#wczytuje plik csv / load the csv file
netflix_df = pd.read_csv("netflix_data.csv")

#pokazuje pierwsze 5 wierszy / display the first 5 row
netflix_df.head()
```

```python
#wyrzucam inne oprócz filmów / filter out everything except movies
netflix_subset = netflix_df[netflix_df["type"]== "Movie"]
netflix_subset
```

```python
#biorę dane od 1990r. w góre / select data from 1990 onwords
subset = netflix_subset[(netflix_subset['release_year']>=1990)]
#odrzucam lata od 2000 w góre / filter out years after 2000
movies_90 = subset[(subset["release_year"]<2000)]

movies_90
#wyświetlam lata 1990-1999 /display the filtered dataset
```

```python
#wizualizuje histogramem po długości filmy / visualize movie duration using a histogram
plt.hist(movies_90["duration"])
plt.title("Distribution of movie duration 1990")
plt.xticks(range(20, 201, 5), rotation=45) #function was used to increase x precision
plt.grid(axis='x', alpha=0.3)
plt.xlabel("Duration (minutes)")
plt.ylabel("Number of movie")
plt.show()
```


* **PL:** Większość filmów z lat 90. mieści się w przedziale **95-111 minut**.
* **EN:** Most movies from the 1990s fall within the **95-111 minute** range.
